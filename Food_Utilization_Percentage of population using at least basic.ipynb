```python
import pandas as pd
import docx
import matplotlib.pyplot as plt
import numpy as np
```

## 1_ Percentage of population using at least basic DRINKING WATER services (percent)

```python
root = '/home/dila-isi/PycharmProjects/homealarms/FAO/FOOD UTILIZATION/'
file = '/Percentage of population using at least basic.csv'

colombia = pd.read_csv(root + 'Colombia' + file)
ecuador = pd.read_csv(root + 'Ecuador' + file)
georgia = pd.read_csv(root + 'Georgia' + file)
peru = pd.read_csv(root + 'Peru' + file)
vietnam = pd.read_csv(root + 'Vietnam' + file)
```

```python
colombia=colombia[colombia['Item']!='Percentage of population using at least basic sanitation services (percent)']
ecuador = ecuador[ecuador['Item']!='Percentage of population using at least basic sanitation services (percent)']
georgia= georgia[georgia['Item']!='Percentage of population using at least basic sanitation services (percent)']
peru = peru[peru['Item']!='Percentage of population using at least basic sanitation services (percent)']
vietnam = vietnam[vietnam['Item']!='Percentage of population using at least basic sanitation services (percent)']
```

```python
colombia
```

```python
colombia.set_index('Year', inplace=True)
ecuador.set_index('Year', inplace=True)
georgia.set_index('Year', inplace=True)
peru.set_index('Year', inplace=True)
vietnam.set_index('Year', inplace=True)
colombia = colombia.tail(10)
ecuador = ecuador.tail(10)
georgia = georgia.tail(10)
peru = peru.tail(10)
vietnam = vietnam.tail(10)
```

```python
dfSummary=pd.DataFrame({"Country":['Colombia', 'Ecuador', 'Georgia','Peru', 'Vietnam' ]})
dfList=[colombia, ecuador, georgia, peru,vietnam ]

min,max,mean,std = [],[],[],[]
for df in dfList:
    min.append("%.2f" % df['Value'].min())
    max.append("%.2f" % df['Value'].max())
    mean.append("%.2f" % df['Value'].mean())
    std.append("%.2f" % df['Value'].std())

dfSummary['Mean']=mean
dfSummary['S.D.']=std
dfSummary['Min.']=min
dfSummary['Max.']=max
```

```python
dfSummary
```

```python
df = dfSummary
# Initialise the Word document
doc = docx.Document()
# Initialise the table
t = doc.add_table(rows=1, cols=df.shape[1])
# Add borders
t.style = 'TableGrid'
# Add the column headings
for j in range(df.shape[1]):
    t.cell(0, j).text = df.columns[j]
# Add the body of the data frame
for i in range(df.shape[0]):
    row = t.add_row()
    for j in range(df.shape[1]):
        cell = df.iat[i, j]
        row.cells[j].text = str(cell)
# Save the Word doc
doc.save('Percentage of population DRINKING WATER services.docx')
```

```python
plt.subplots(figsize=(10, 5))
plt.plot(colombia['Value'], label='Colombia', color='green', marker='o')
plt.plot(ecuador['Value'], label='Ecuador', color='steelblue', marker='o')
plt.plot(georgia['Value'], label='Georgia', color='purple', marker='o')
plt.plot(peru['Value'], label='Peru', color='black', marker='o')
plt.plot(vietnam['Value'], label='Vietnam', color='red', marker='o')
plt.title('Percentage of population using at least basic drinking water services (percent)', fontsize=12)
plt.legend(title='Country')
plt.ylim(87,98)
plt.ylabel('%', fontsize=14)
#plt.figure(figsize=(15,100))


#plt.plot(range(10), linestyle='--', marker='o', color='b', label='line with marker')
plt.legend()
```
